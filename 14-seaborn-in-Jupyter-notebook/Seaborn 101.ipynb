## Seaborn 

Relplot = Relational Plot (To get the relationship between two numeric columns),
Catplot = Category Plot (To get the relationship between one numeric column and one non-numeric column),
Distplot = Distribution Plot (To get the distribution of one numeric column)

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
sb
```

```python
tips = sb.load_dataset('tips') # Tips is a in built dataset by seaborn
```

```python
tips
```

```python
sb.relplot(x="total_bill",y="tip",data=tips) # data=tips is the dataframe name, This below chart shows the relationship between total_bill and tip
```

```python
sb.relplot(x="total_bill",y="tip",data=tips,col="smoker") # We are adding col="smoker" and we can see two charts one for smoker=yes and other for smoker=no
```

```python
sb.relplot(x="total_bill",y="tip",data=tips,hue="smoker") # We are adding hue="smoker" to see the same data of smoker in single chart
```

```python
sb.relplot(x="total_bill",y="tip",data=tips,hue="smoker",col="time") # Adding new column = "time" 
```

```python
sb.relplot(x="total_bill",y="tip",data=tips,hue="smoker",row="day",col="time") # Adding row = "day"
```

```python
sb.relplot(x="day",y="tip",data=tips,kind="line") #  kind tells the type of plot we need, by default it is scatter, here we are setting line 
```

```python
sb.catplot(x="day",y="tip",data=tips) # This is a category plot, we are plotting the tip by day, This is also called "Strip plot"
```

```python
sb.catplot(x="smoker",y="tip",data=tips) # This shows the relationship between smoker and tip
```

```python
sb.catplot(x="time",y="tip",data=tips)  # This shows the relationship between time and tip
```

```python
sb.catplot(x="time",y="tip",data=tips,jitter=False)  # This is showing the same data as above plot but, in unscattered manner
```

```python
sb.catplot(x="time",y="tip",data=tips,kind="swarm") # This is showing each data points clearly, not overlapping like scatter
```

```python
sb.catplot(x="day",y="tip",data=tips,kind="bar") # This is showing avg tip by day, The black line at the top is the confidence level
```

```python
sb.catplot(x="day",y="tip",data=tips,kind="bar",hue="time") # Adding new column time
```

```python
sb.catplot(x="day",y="tip",data=tips,kind="box") # Below is the box plot, showing tip by day
```

```python
sb.catplot(y="tip",data=tips,kind="box") # Below is the box plot, showing tip regardless of the day, as we have not used x-axis value
```

```python
sb.catplot(x="day",data=tips,kind="count") # Getting the count of tips received by day
```

```python
sb.catplot(x="day",data=tips,kind="count",hue="smoker") # Getting the count of tips received by day by smoker
```

```python
tips["smoker"].value_counts() # Chaecking the count
```

```python
sb.catplot(x="day",y="tip",data=tips,kind="point") # Point plot, dwawing the tip by day
```

```python
sb.catplot(x="day",y="tip",data=tips,kind="point",hue="smoker") # Point plot, dwawing the tip by day and by smoker
```

```python
sb.__version__ # We can use __version__ to check the version of any lib
```

```python
sb.distplot(tips["tip"]) # This is distrubution plot, showing the distribution of tip 
```

```python
sb.distplot(tips["tip"],bins=range(0,12,1)) # This is distrubution plot, showing the distribution of tip 
plt.xticks(range(0,12,1)) # changing the ticks to 1 for xaxis
```

```python
fig = plt.figure(figsize=(15,10)) # This creates a container with 30 is the width of the bar and 10 height of the bar
#This is creating 2*2 matrix to plot 4 charts  
fig.add_subplot(2,2,1)
plt.bar(tips["day"],tips["tip"])
fig.add_subplot(2,2,2)
plt.bar(tips["day"],tips["tip"])
fig.add_subplot(2,2,3)
plt.bar(tips["day"],tips["tip"])
fig.add_subplot(2,2,4)
plt.bar(tips["day"],tips["tip"])
```

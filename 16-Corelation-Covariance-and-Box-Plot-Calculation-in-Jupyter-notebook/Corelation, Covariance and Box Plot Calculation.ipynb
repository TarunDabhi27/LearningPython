### Corelation
Corelation varies from -1 ro +1

Positive Corelation 0.1 to 0.3 | weak | -0.1 to 0.3 Negative Corelation <br>
Positive Corelation 0.3 to 0.5 | medium | -0.3 to 0.5 Negative Corelation <br>
Positive Corelation 0.5 to 1 | high | -0.5 to 1 Negative Corelation

### Getting the value for Box Plot

```python
Q1 = np.percentile(bike_sharing_data["temp"],25) # quartile 1
```

```python
Q1, Q2, Q3
```

```python
Q2 = np.percentile(bike_sharing_data["temp"],50) # quartile 2
```

```python
Q3 = np.percentile(bike_sharing_data["temp"],75) # quartile 3
```

```python
IQR = Q3 - Q1 # Inter Quartile Range
```

```python
IQR
```

```python
lower_bound = Q1 - (1.5*IQR) # Getting the lower bound, beyond this all points are possible ouliers
```

```python
lower_bound
```

```python
upper_bound = Q3 + (1.5*IQR) # Getting the upper bound, beyond this all points are possible ouliers
```

```python
upper_bound
```

```python
bike_sharing_data[bike_sharing_data["temp"]>upper_bound] # Outliers which are higher then upper_bound, In this scenation we do not have any outlier
```

```python
bike_sharing_data[bike_sharing_data["temp"]<lower_bound] # Outliers which are higher then lower_bound, In this scenation we do not have any outlier
```

```python
sb.catplot(x="temp",data=bike_sharing_data,kind="box")
```

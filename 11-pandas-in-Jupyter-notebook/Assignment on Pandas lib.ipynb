```python
bike_sharing_data.groupby("holiday").agg({"cnt":"count"}) # Count by holiday
```

```python
bike_sharing_data[(bike_sharing_data["temp"]>.2) & (bike_sharing_data["hum"]>.2)] # This will give data where temp and hum is > .2
```

```python
bike_sharing_data.dropna(how="any",axis=0) # This will drop the row where any value is null
```

```python
bike_sharing_data.dropna(how="all",axis=0) # This will drop the row where all value is null
```

```python
bike_sharing_data.drop("atemp",axis=1) # This will drop the atemp column
```

```python
bike_sharing_data.groupby("season").agg({"temp":["max","min"]}) # This will give max and min temp season wise
```

```python
bike_sharing_data.drop(bike_sharing_data[bike_sharing_data["windspeed"]>0.2].index, axis=0) # This will drop the rows where windspeed > 0.2
```
